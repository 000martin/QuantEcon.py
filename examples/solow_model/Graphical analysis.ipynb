##3.2  Intensive investment
The steady state, or long-run equilibrium, of the Solow growth model balances actual investment (per unit effective labor) with effective depreciation. We can study the impact of various parameters on the long-run equilibrium value of capital (per unit effective labor) as follows...

```python
def intensive_output_widget(model, **params):
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    model.plot_intensive_output(ax, Nk=1000, **params)
```

```python
# define some widgets for the various parameters
eps = 1e-2
technology_progress_widget = FloatSliderWidget(min=-0.05, max=0.05, step=eps, value=0.0)
population_growth_widget = FloatSliderWidget(min=-0.05, max=0.05, step=eps, value=0.0)
savings_widget = FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.5)
output_elasticity_widget = FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.5)
depreciation_widget = FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.5)
```

```python
# create the interactive plot
interact(intensive_output_widget,
         model=fixed(model),
         g=technology_progress_widget,
         n=population_growth_widget,
         s=savings_widget, 
         alpha=output_elasticity_widget,
         delta=depreciation_widget,
         )
```

## 3.3 Solow diagram

```python
def solow_diagram_widget(model, **params):
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))
    model.plot_solow_diagram(ax, Nk=1000, **params)
    
# define some widgets for the various parameters
eps = 1e-2
technology_progress_widget = FloatSliderWidget(min=-0.05, max=0.05, step=eps, value=0.0)
population_growth_widget = FloatSliderWidget(min=-0.05, max=0.05, step=eps, value=0.0)
savings_widget = FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.5)
output_elasticity_widget = FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.5)
depreciation_widget = FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.5)

# create the widget!
interact(solow_diagram_widget, 
         model=fixed(model),
         g=technology_progress_widget,
         n=population_growth_widget,
         s=savings_widget, 
         alpha=output_elasticity_widget,
         delta=depreciation_widget,
         )
```

## 3.4 Phase Diagram
So long as the initial levels of capital, labor, and technology are all strictly positive, the Solow model is globally stable: given any initial condition for capital per effective worker, $k$, the dynamics of the Solow model will push $k$ towards its long-run, steady state value, $k^*$. This information can be summarized graphically by a phase diagram.  The `solow.Model` class has a method for plotting the phase diagram of the model...

```python
# define some widgets for the various parameters
eps = 1e-2
technology_progress_widget = FloatSliderWidget(min=-0.05, max=0.05, step=eps, value=0.0)
population_growth_widget = FloatSliderWidget(min=-0.05, max=0.05, step=eps, value=0.0)
savings_widget = FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.5)
output_elasticity_widget = FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.5)
depreciation_widget = FloatSliderWidget(min=eps, max=1-eps, step=eps, value=0.5)

# create the widget!
phase_diagram_widget = interact(solow.model.plot_phase_diagram, 
                                cls=fixed(model),
                                Nk=fixed(value=1e3),
                                g=technology_progress_widget,
                                n=population_growth_widget,
                                s=savings_widget, 
                                alpha=output_elasticity_widget,
                                delta=depreciation_widget,
                                )
```
