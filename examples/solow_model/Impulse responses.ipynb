# 5. Impulse response functions

Impulse response functions are a standard tool for analyzing the short run dynamics of dynamic macroeconomic models. The `solow.Model` class has an `irf` attribute that stores information about various impulse responses...

### Impact of a change in the savings rate

```python
# 5% increase in the current savings rate...
model.irf.impulse = {'s': 1.05 * model.params['s']}

# in per capita units...
model.irf.kind = 'per_capita'
```

Take a look at the IRF for the savings rate shock. Note that while capital and output are unaffected at the `t=0`, both consumption and investment jump (in opposite directions!) in response to the change in the savings rate.

```python
# take a look at the impulse reponse
print(model.irf.impulse_response[:25, ])
```

```python
fig, ax = solow.impulse_response.plot_impulse_response(model,
                                                       variable='output',
                                                       impulse={'s': 1.05 * model.params['s']},
                                                       kind='efficiency_units',
                                                       )
```

Want to allow for fairly general shocks. For example, suppose that an exogenous 5% increase in the growth rate of technology was accompanied by a simultaneous 10% fall in the savings rate.

```python
shock = {'s': 0.9 * model.params['s'], 'g': 1.05 * model.params['g']}

fig, ax = solow.impulse_response.plot_impulse_response(model,
                                                       variable='consumption',
                                                       impulse=shock,
                                                       kind='per_capita',
                                                       log=True,
                                                       )
```
