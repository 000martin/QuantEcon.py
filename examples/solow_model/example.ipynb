# Motivation
Most instructors teach the Solow model to undergraduates (and graduates!) using the Cobb-Douglas functional form for production. However Cobb-Douglas functional form predicts constant factor shares. Constant factor shares are inconsistent with post-war experience for most countries...

```python
pwt = pypwt.load_pwt_data()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,6))

for ctry in pwt.major_axis:
    tmp_data = pwt.major_xs(ctry)
    tmp_data.labsh.plot(color='gray', alpha=0.5)
    
# plot some specific countries
pwt.major_xs('USA').labsh.plot(color='blue', ax=ax, label='USA')
pwt.major_xs('IND').labsh.plot(color='green', ax=ax, label='IND')
pwt.major_xs('CHN').labsh.plot(color='orange', ax=ax, label='CHN')

# plot global average
avg_labor_share = pwt.labsh.mean(axis=0)
avg_labor_share.plot(color='r', ax=ax)

ax.set_xlabel('Year', family='serif', fontsize=15)
ax.set_ylabel('Labor share of income', family='serif', fontsize=15)
ax.set_ylim(0, 1)

plt.show()
```

Note that variable factor shares are inconsistent with the notion of a balanced growth path! Thus trends in factor shares that we observe in the data imply that we must focus the analysis on the disequilibrium dyanmics when seeking to validate the Solow model.
